# DriveEase Project Directory Interaction

This notebook demonstrates how to interact with the already created `mydrivease` directory. It covers verifying the directory, listing its contents, creating new files, and reading/writing to files.

## 1. Verify Existing Directory
Check if the 'mydrivease' directory exists using Python's os module.

```python
import os

# Path to check
path = 'mydrivease'

# Check if directory exists
if os.path.isdir(path):
    print(f"Directory '{path}' exists.")
else:
    print(f"Directory '{path}' does NOT exist.")
```

## 2. List Files in Directory
List all files and subdirectories inside 'mydrivease' using os.listdir().

```python
# List all files and subdirectories in 'mydrivease'
if os.path.isdir(path):
    print(os.listdir(path))
else:
    print(f"Directory '{path}' does NOT exist.")
```

## 3. Create New Files in Directory
Create new files within the 'mydrivease' directory using open() in write mode.

```python
# Create a new file in 'mydrivease' directory
new_file_path = os.path.join(path, 'example.txt')
with open(new_file_path, 'w') as f:
    f.write('This is a test file in mydrivease directory.')
print(f"Created file: {new_file_path}")
```

## 4. Read and Write to Files in Directory
Demonstrate reading from and writing to files located in 'mydrivease'.

```python
# Write to the file
with open(new_file_path, 'a') as f:
    f.write('\nAdding another line to the file.')

# Read from the file
with open(new_file_path, 'r') as f:
    content = f.read()
print('File content:')
print(content)
```
